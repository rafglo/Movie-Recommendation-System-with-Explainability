```python
import pandas as pd
```

```python
raw_dir = '../data/raw'
processed_dir = '../data/processed'
ratings = pd.read_csv(f'{raw_dir}/ratings.csv')
movies = pd.read_csv(f'{raw_dir}/movies.csv')
```

Downsampling

```python
print(f"Original ratings: {len(ratings)}")

# 1. Keep only users who have rated at least 500 movies (The "Super Users")
user_counts = ratings['userId'].value_counts()
active_users = user_counts[user_counts >= 500].index
ratings = ratings[ratings['userId'].isin(active_users)]

# 2. Keep only movies that have at least 500 ratings (The "Blockbusters")
movie_counts = ratings['movieId'].value_counts()
popular_movies = movie_counts[movie_counts >= 500].index
ratings = ratings[ratings['movieId'].isin(popular_movies)]

print(f"Safe downsampled ratings for laptop RAM: {len(ratings)}")
```

```python
ratings['datetime'] = pd.to_datetime(ratings['timestamp'], unit='s')

# Merge to create the master file
master_df = ratings.merge(movies[['movieId', 'title']], on='movieId', how='left')

# Save to parquet for lightning-fast loading in our other notebooks
master_df.to_parquet(f'{processed_dir}/master_data_v2.parquet')
```
